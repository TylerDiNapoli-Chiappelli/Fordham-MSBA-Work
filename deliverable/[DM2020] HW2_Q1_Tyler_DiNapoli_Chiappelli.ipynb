```python
import pandas as pd
import seaborn as sb
import numpy as np

#insheet the data
data = pd.read_csv('HW2_Cereals.csv')

#Question 1a: compare "shelf" to "rating". Separate data by shelf and then box plot against rating
y_var = 'rating'
x_var = 'shelf'

boxplot_graph = data.boxplot(column = y_var, by = x_var)
print (boxplot_graph)
```

```python
#Question 1b: make a heat map
correlation_matrix = data.corr()
heatmap = sb.heatmap(correlation_matrix, xticklabels = correlation_matrix.columns, yticklabels = correlation_matrix.columns)
print (heatmap)
```
